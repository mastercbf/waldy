```python
import pandas as pd
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
df2 = table[1]
df.info()
keep = df[['Symbol','GICS Sector']]
sp500 = keep.rename(columns = {'GICS Sector':'Industry'})
sp500['Symbol']
stock_tickers = sp500['Symbol'].values.tolist()
stocks = ["SWTSX","NDXT","AAPL","MSFT","GOOG","AMZN","NVDA","TSLA","META","BABA","CRM"]  + stock_tickers
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine import *
import yfinance as yf


# IMPORT THE LIBRARY
from datetime import datetime
end_date = datetime.now().strftime('%Y-%m-%d')
i=0

test = pd.DataFrame()
for stock in stocks:
  

  stocks[i] = yf.Ticker(stock)
  stocks[i]= stocks[i].history(start='2023-01-01',end=end_date)
  ticker = stock
  #print(stocks[i])
  stocks[i]['Ticker'] = ticker
  #print(stocks[i])
  #print(stocks[0])
  test = test.append(stocks[i])
  i+=1
test['win_loss_value'] = test.Close - test.Open
test['win_loss'] = np.where(test['win_loss_value']>0, 'Up', 'Down')
test.reset_index(inplace=True)
daily_stocks = test
daily_stocks.info()
```

# Merging Final Dataset to include Industry



```python
daily_stock_final = daily_stocks.merge(sp500 , how = 'left', left_on='Ticker',right_on='Symbol')

daily_stock_final['Ticker']
#['SWTSX']
#print(daily_stock_final)
daily_stock_final.head()

#d#aily_stock_final.assign(Ticker."SWTSX" = "Vandelay Industries")


```
